<h2>Simulation of Protein Synthesis and Degradation without Tellurium</h2>


```python
import numpy as np
import matplotlib.pyplot as plt

k0 = 0.01
k1 = 1
k2 = 5
mRNA = 5
pro_conc = 0
n_points = 100

mRNA_conc = np.linspace(mRNA, 0, n_points)
steady_state_sol = (k0+k1*mRNA_conc) / k2
Rate = k0+(k1*mRNA_conc) - (k2* pro_conc)

plt.figure(figsize=(10, 7))
plt.rcParams['axes.facecolor'] = 'lightgray'
plt.plot(steady_state_sol, Rate, label='Protein Responce', color='red')
plt.scatter(steady_state_sol[np.abs(Rate - 1).argmin()], 1, color='blue', marker='o')
plt.scatter(steady_state_sol[np.abs(Rate - 2).argmin()], 2, color='blue', marker='o')
plt.scatter(steady_state_sol[np.abs(Rate - 3).argmin()], 3, color='blue', marker='o')
plt.axhline(y=1, color='blue', linestyle='--',label= 'Signal(mRNA) Concentration')
plt.axhline(y=2, color='blue', linestyle='--')
plt.axhline(y=3, color='blue', linestyle='--')
plt.xlim(0, 1)  
plt.ylim(0, 5) 
plt.xlabel('Response (Protein Concentration)',fontsize=12)
plt.ylabel('Rate(dR/dt)',fontsize=12)
plt.title('The Rate Curve of Protein Synthesis & Degradation',fontsize=16)
plt.legend()
plt.grid(False)
plt.show()
```

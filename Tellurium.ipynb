```python
import tellurium as te

r = te.loada("""
         J1: a -> b; k1*a
         J2: b -> c; k2*b
         J3: b + c -> a; k3*c
         
         a = 100;
         b = 60;
         c = 10;
         
         k1 = 2;
         k2 = 3;
         k3 = 1;
         
         
         
""")

result = r.simulate(0, 10, 100)
r.plot()





```

```python
import tellurium as te
import numpy as np
import matplotlib.pyplot as plt

def goldbeter_koshland(u, v, J, K):
    numerator = 2 * u * K
    denominator = v - u + v * J + u * K * np.sqrt((v - u + v * J + u * K)**2 - 4 * (v - u) * u * K)
    return numerator / denominator

def steady_state_concentration(S, k1, k2, km1, km2, Rt):
    J = km1 / Rt
    K = km2 / Rt
    u = k1 * S
    v = k2
    G = goldbeter_koshland(u, v, J, K)
    return G * Rt

k1 = 1.0
k2 = 1.0
km1 = 0.05
km2 = 0.05
Rt = 1.0

S_values = np.linspace(0, 3, 100)
Rp_ss_values = [steady_state_concentration(S, k1, k2, km1, km2, Rt) for S in S_values]

plt.plot(S_values, Rp_ss_values)
plt.xlabel('S (Signal)')
plt.ylabel('Rp,ss (Steady-state concentration of phosphorylated protein)')
plt.title('Sigmoidal Signal-Response Curve')
```

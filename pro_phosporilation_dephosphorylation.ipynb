<h2>Simulation of Phosphorylation and Dephosphorylation of Protein without Tellurium</h2>

```python
import numpy as np
import matplotlib.pyplot as plt

k1 = 1
k2 = 1
Rt = 1
Rp = 0
Rt_conc = np.linspace(Rt, 0, 100)
Rp_conc = np.linspace(Rt, 0, 100)
dPro_dTime = (k1* Rt_conc)*(Rt - Rp) - (k2*Rp)


plt.figure(figsize=(10, 7)) 
plt.rcParams['axes.facecolor'] = 'lightgray'
plt.plot(Rp_conc, dPro_dTime, color='red', label='Protein Responce')
plt.plot([1, 0], [0, 2], 'blue', linestyle='--', label='Signal Concentration')  
plt.plot([1, 0], [0, 4], 'blue', linestyle='--')  
plt.plot([1, 0], [0, 8], 'blue', linestyle='--')
plt.scatter(0.668, 0.669, color='blue')  
plt.scatter(.8, .8, color='blue')  
plt.scatter(.89, .89, color='blue')
plt.xlabel('Concentration of Phosporilaited Protein(Rp)', fontsize=12)
plt.ylabel('Rate (dRp/dt)', fontsize=12)
plt.title('The Rate Curve of Protein Phosphorylation', fontsize=16)
plt.ylim(0,2)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

k1 = 1
k2 = 1
Rt = 1
Rp = 0

S_conc = np.linspace(0,10,100)

Rp_ss = (Rt * S_conc) / ((k2/k1) + S_conc) 



plt.figure(figsize=(10, 7))  
plt.rcParams['axes.facecolor'] = 'lightgray'
plt.plot(S_conc, Rp_ss, color='red')
plt.xlabel('Concentration of Signal', fontsize=12)
plt.ylabel('The Steady_state Response(Rp_ss)', fontsize=12)
plt.title('The Signal-Response Curve of Protein Phosphorylation', fontsize=16)
plt.grid(True)
plt.show()
```
